```python
from tensorflow.keras.layers import LSTM, RNN, Input, Dense,Conv1D,MaxPooling1D,Flatten,BatchNormalization
from tensorflow.keras.models import Model

step = 200
n_inputs = 42
n_hiddens = 64
n_outputs = 1
inputs = Input(shape=(step, n_inputs))
x = Conv1D(256, 10, activation='relu')(inputs)
x = MaxPooling1D(5)(x)
x = Flatten()(x)
x = Dense(32,activation='relu')(x)
output19 = Dense(n_outputs)(x)
output20 = Dense(n_outputs)(x)
output21 = Dense(n_outputs)(x)
output22 = Dense(n_outputs)(x)
output23 = Dense(n_outputs)(x)
output24 = Dense(n_outputs)(x)

model = Model(inputs,[output19,output20,output21,output22,output23,output24])
model.summary()
```

```python
model.compile(optimizer='adam', loss='mse',metrics=['mae'])
history = model.fit(X_train,[Y_train19,Y_train20,Y_train21,Y_train22,Y_train23,Y_train24], epochs= 200, batch_size=128)
model.save("home/iotadmin/notebooks/model/right_right_multitask.h5")
```
