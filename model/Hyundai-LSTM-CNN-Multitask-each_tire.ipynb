```python
from tensorflow.keras.layers import LSTM, RNN, Input, Dense,Conv1D,MaxPooling1D,Flatten,BatchNormalization
from tensorflow.keras.models import Model

step = 200
n_inputs = 24
n_hiddens = 256
n_outputs = 1

inputs = Input(shape=(step, n_inputs))
x = Conv1D(32, 5, activation='relu')(inputs)
x = MaxPooling1D(5)(x)
x = Conv1D(32, 5, activation='relu')(x)
x = MaxPooling1D(5)(x)
x = Conv1D(32, 5, activation=
           'relu')(x)
x = MaxPooling1D(3)(x)
#hidden3 = LSTM(n_hiddens)(hidden2)
#hidden_d = Dense(128,activation='relu')(hidden3)
#hidden_d1 = Dense(64,activation='relu')(hidden_d)
#hidden_d1 = Dense(128,activation='relu')(hidden2)
#hidden_d2 = Dense(64,activation='relu')(hidden_d1)
#hidden_d3 = Dense(32,activation='relu')(hidden_d2)
x = LSTM(n_hiddens,activation='relu')(x)
#hidden3 = LSTM(n_hiddens)(hidden2)
#hidden_d = Dense(128,activation='relu')(hidden3)
#hidden_d1 = Dense(64,activation='relu')(hidden_d)

x = Dense(128,activation='relu')(x)
x = Dense(64,activation='relu')(x)
x = Dense(32,activation='relu')(x)
output19 = Dense(n_outputs)(x)
output20 = Dense(n_outputs)(x)
output21 = Dense(n_outputs)(x)
output22 = Dense(n_outputs)(x)
output23 = Dense(n_outputs)(x)
output24 = Dense(n_outputs)(x)


model = Model(inputs,[output19,output20,output21,output22,output23,output24])
model.summary()
```
