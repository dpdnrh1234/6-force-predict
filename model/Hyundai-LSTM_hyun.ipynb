```python
import pandas as pd
import numpy as np
from tqdm import tqdm
import random
import tensorflow as tf
import matplotlib.pyplot as plt
import os
```

```python
#Input 전체
#X_train = np.load('/home/iotadmin/notebooks/array/window_50_t_train_split.npy')

#Input 해당
X_train = np.load('/home/iotadmin/notebooks/array/window_200_t.npy')
```

```python
Y_train = np.load('/home/iotadmin/notebooks/array/window_200_t_4.npy')
```

```python


#Input 해당
X_test = np.load('/home/iotadmin/notebooks/array/test_window_200_s.npy')
```

```python
Y_test = np.load('/home/iotadmin/notebooks/array/test_window_200_t_4.npy')
```

```python
from tensorflow.keras.layers import LSTM, RNN, Input, Dense,Conv1D,MaxPooling1D,Flatten,BatchNormalization,Flatten
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Bidirectional

step = 200
n_inputs = 42
n_hiddens = 64
n_outputs = 1


inputs = Input(shape=(step, n_inputs))
x = Conv1D(32, 5, activation='relu')(inputs)
x = MaxPooling1D(5)(x)
x = Conv1D(32, 5, activation='relu')(x)
x = MaxPooling1D(5)(x)
x = Conv1D(32, 5, activation='relu')(x)
x = MaxPooling1D(3)(x)

x = LSTM(n_hiddens,return_sequences=True)(x)
x = LSTM(n_hiddens)(x)

output = Dense(n_outputs)(x)

model = Model(inputs,output)
model.summary()
```

```python
model.compile(optimizer='adam', loss='mse',metrics=['mae'])
history = model.fit(X_train, Y_train, epochs= 200, batch_size=128)
model.save("home/iotadmin/notebooks/model/FLMX_CNN_LSTM_last.h5")
```

```python
from tensorflow.keras.models import load_model
model = load_model("home/iotadmin/notebooks/model/FLMX_CNN_LSTM_last.h5")
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
Y_predict = model.predict(X_test)
Y_predict=Y_predict.reshape(Y_predict.shape[0],)
print(mean_squared_error(Y_test, Y_predict)**0.5)
print((np.corrcoef([Y_test,Y_predict])[0][1])**2)
```
